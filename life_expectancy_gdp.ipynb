```python
# Import the appropriate libraries before beginning.

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Print the first five rows of the dataframe.

df = pd.read_csv('all_data.csv')
print(df.head())
```

```python
# Print the information and description about the dataframe.

print(df.info())
print(df.describe())
```

```python
# Find out more about each column in the dataframe.

print(df.Country.value_counts())
print(df.Year.value_counts())
print(df["Life expectancy at birth (years)"].value_counts())
print(df.GDP.value_counts())
```

```python
# What is the life expectancy for each country over time?

sns.set_style('darkgrid')
sns.set_palette("colorblind")

f, ax = plt.subplots(figsize=(10,6))
sns.lineplot(data=df, x='Year', y='Life expectancy at birth (years)', hue='Country')

plt.show()
plt.clf()
```

```python
# Find the average life expectancy for each country.

f, ax = plt.subplots(figsize=(10,6))
sns.barplot(data=df, x='Country', y='Life expectancy at birth (years)', order=['Germany','Chile','United States of America','Mexico','China','Zimbabwe'])
```

```python
# What is the GDP among each country over time?

sns.set_style('darkgrid')
sns.set_palette('colorblind')

f, ax = plt.subplots(figsize=(10,6))
sns.lineplot(data=df, x='Year', y='GDP', hue='Country')

plt.show()
plt.clf()
```

```python
# Is there a correlation between a country's GDP and its life expectancy?

countries = df.Country.unique()
df['GDP in Trillions'] = df['GDP'] / 1000000000000

f = plt.figure(figsize=(10,10))
ax = plt.subplot(3,2,6)

for i, country in zip(range(1, 7), countries):
    axi = plt.subplot(3,2,i)
    sns.scatterplot (data=df[df.Country == country], x='GDP in Trillions', y='Life expectancy at birth (years)')
    plt.title('Life Expectancy vs GDP for {}'.format(country))
    
plt.subplots_adjust(wspace=.3, hspace=.4)
plt.show()
plt.clf()
```
